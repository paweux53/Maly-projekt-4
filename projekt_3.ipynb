## 5. Analiza przekroczeń normy wg województw

Dodatkowo, sprawdzimy jak wygląda liczba dni z przekroczeniem normy PM2.5 w podziale na województwa. W tym celu mapujemy stacje pomiarowe do odpowiednich województw na podstawie metadanych. Wykres przedstawia średnią liczbę dni z przekroczeniem normy dla stacji w danym województwie.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Mapowanie
meta = pd.read_excel('metadane_new.xlsx', index_col=0)
col_kod = next((c for c in meta.columns if 'Kod' in str(c)), 'Kod') 
col_woj = next((c for c in meta.columns if 'Województwo' in str(c)), 'Województwo')
mapping = dict(zip(meta[col_kod], meta[col_woj]))

# Obliczenia przekroczeń (> 15 ug/m3)
res = {}
years = sorted(df.index.year.unique())

for y in years:
    sub = df[df.index.year == y]
    # Średnie dobowe
    daily = sub.groupby(sub.index.date).mean()
    # Liczba dni z przekroczeniem
    count = (daily > 15).sum()
    res[y] = count

# Przetwarzanie i wizualizacja po pętli (tylko raz)
df_res = pd.DataFrame(res)

# Dodanie kolumny Województwo
df_res['Województwo'] = df_res.index.get_level_values(0).map(mapping)

# Grupowanie po województwach (średnia dla stacji w województwie)
df_woj = df_res.groupby('Województwo').mean()

# Wizualizacja
df_long = df_woj.reset_index().melt(id_vars='Województwo', var_name='Rok', value_name='Dni')

sns.set_style("whitegrid")
plt.figure(figsize=(14, 7))
sns.barplot(data=df_long, x='Województwo', y='Dni', hue='Rok')
plt.title("Średnia liczba dni z przekroczeniem normy PM2.5 (15 µg/m³) w województwach")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Średnia liczba dni")
plt.legend(title='Rok')
plt.tight_layout()

plt.show()
```

## Interpretacja
W 2024 najmniejszą liczbę dni przekroczeń normy PM2.5 miały województwa Podlaskie i Zachodniopomorskie. Jest to zmiana w porównianiu z 2021, kiedy najlepiej wypadało województwo Lubuskie, które w tamtym roku zanotwało wyjątkowo niską liczbę dni przekroczeń normy.

Negatywnie natomiast wyrónia się województwo Małopolskie, które systematycznie notuje wysokie przekroczenia normy PM2.5 w ciągu ostatnich lat.

Pozytywnym aspektem jest wyraźny spadek PM2.5 na przestrzeni lat we wszystkich województwach.
